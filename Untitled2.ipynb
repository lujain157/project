```python
#import lipraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
td = pd.read_csv('/content/StudentsPerformance.csv')
```

```python
td.head()
```

```python
sns.heatmap(td.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
td.drop((['parental level of education','lunch','test preparation course']),axis=1,inplace=True)
```

```python
td.head()
```

```python
gendar=pd.get_dummies(td['gender'],drop_first=True)
```

```python
td.drop(['gender'],axis=1,inplace=True)
```

```python
td = pd.concat([td,gendar],axis=1)
```

```python
td.head()
```

```python
ethnicity = pd.get_dummies(td['race/ethnicity'], prefix='ethnicity', drop_first=True)
```

```python
td = pd.concat([td, ethnicity], axis=1)
```

```python
td = td.drop('race/ethnicity', axis=1)
```
